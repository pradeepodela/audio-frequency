```python
from recorder import Recorder
import matplotlib.pyplot as plt
import scipy
import librosa
from librosa import display
from IPython.display import Audio
from scipy.fft import fft, fftfreq
import numpy as np
```

```python

r = Recorder()
r.record(5, output='out.wav')
fname="out.wav"
samples, sam_rate=librosa.load(fname, sr=None, mono=True, offset=0.0, duration=None)


Audio(fname)

plt.figure()
librosa.display.waveplot(y=samples, sr=sam_rate)
plt.xlabel("time secs")
plt.ylabel("Ampl")
plt.show()

def fft_plot(audio, sam_rate):
    n = len(audio)
    T= 1/sam_rate
    yf = fft(audio)
    xf = fftfreq(n, T)#[:n//2]
    fig, ax = plt.subplots()
    plt.plot(xf, 2.0/n * np.abs(yf)) #np.abs(yf[0:n//2]))
    plt.grid()
    plt.xlabel("Freq")
    plt.ylabel("mag")
    val=np.argmax(yf)
    plt.show()
    return np.abs(xf[val])


audio_freq = fft_plot(samples, sam_rate)
print("Max frequency is",audio_freq.round())
```
